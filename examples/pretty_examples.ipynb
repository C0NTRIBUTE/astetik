```python
%matplotlib inline
import pretty as pretty
import pandas as pd
```

# 1. Overview of Pretty's functionality

First let's get some data (in this case 1000 tweets about Donald Trump. I'm using Somecode Twitter Research kit to get the data directly in to a pandas dataframe from Twitter API

```python
import somecode as some
df = some.search("trump",100)
```

## 1.1. List of Functions

- Text (e.g. tweeets) 
- Descriptive Stats
- Bars
- Side-by-side Histogram
- Bubble Chart (4 dimensions)
- Heatmap Correlation (up to 40 variables)
- Kernel Density Estimation
- Swarming 
- OLS Regression test

## 1.2. Common function parameters

Because generally the user is required to spend much time to figure out basic configuration, such as titles, scales etc, Pretty is focused on making frequently repeated operations as intuitive accessible as possible. 

Example paramaters: 

- title / string to be used as a title for the graphic
- suptitle / string to be used as subtitle for the graphic
- xscale / for setting the scale of x axis ("linear","symlog","log")
- yscale / see above line 
- color / for changing a single color
- palette / for changing a palette of colors (Pretty is using Seaborn palettes but you could use any)

# 2. Examples of text and table presentation

The below examples highlight Pretty's functionality when used at the most basic level.

## 2.1. Tweets with most retweets

```python
pretty.text(df,'text',5,sort_by='retweet_count')
```

## 2.2. Negative tweets sorted by number of retweets

```python
pretty.text(df[df.neg > 0.2],'text',"Negative tweets with many retweets",5,sort_by='retweet_count')
```

## Descritive statistics in a table

```python
pretty.descriptive(df,['neg','neu','pos'],"Sentiment")
```

## OLS Regression Test

```python
pretty.ols(df,'retweet_count','neg','neu','pos')
```

# 3. Examples of plot presentation

## 3.1. Side-by-side histograms 

```python
pretty.histogram(df,['quality_score','neu'])
```

## 3.2. Horizontal bars for a single dimension of data

```python
pretty.bars(df.location)
```

## 3.3. Bubble chart for displaying 4 dimensions of data

```python
pretty.bubble(df,'influence_score','reach_score','egg_account','retweet_count',xscale='symlog', yscale='symlog')
```

## 3.4. Correlation heatmap for up to 30 variables 

```python
pretty.correlationt(df)
```

## 3.5. Kernel Density Estimation

```python
pretty.kde(df.influence_score,df.retweet_count,xscale="symlog",yscale="symlog")
```

## 3.6. Swarming

```python
pretty.swarm(df,'neg','retweet_count',yscale="symlog")
```
